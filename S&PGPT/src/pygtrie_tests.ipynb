```python
import pandas as pd
import pygtrie
```

```python
df = pd.read_csv("../data/nasdaq_listed_companies.csv")
```

```python
df.head()
```

```python
df['clean_name'] = df['name'].str.replace(r'\s+(Common Stock|Units|Common Stock |Units )$', '', regex=True)
```

```python
df['Response'] = df['symbol'] + ' - ' + df['clean_name']

df.head()
```

```python
# Create a Trie data structure
trie = pygtrie.CharTrie()

# Insert symbols into the Trie
for symbol in df['Response'].astype(str):
    trie[symbol] = True

print(list(trie.keys(prefix='MSFT')))
print("Trie data structure has been created successfully.")
```
